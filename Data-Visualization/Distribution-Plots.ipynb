```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
tips_df = sns.load_dataset('tips')
tips_df
```

## Distribution plots are used to visualize the distribution of `quantitative data`.
## Some of these plots are:
* ### `distplot`/`histplot`
* ### `jointplot`
* ### `pairplot`
* ### `rugplot`

## 1. `histplot`/`distplot`:
* ### `histplot` is basically a frequency table.
* ### It counts the frequency of the values that are landed in `bins`
* ### `Bins` are a kind of continous range like 0-10, 10-20, 20-30 etc... 
* ### If any value lands in one of these bins the count of that `bin` increases.
* ### `kde` or kernel density estimator is the kwarg of `histplot` and it draws a fitting curve for the plot.

```python
plt.subplot(1, 2, 1)
sns.histplot(data=tips_df, x='total_bill', bins = 20, kde = True)

plt.subplot(1,2,2)
sns.histplot(data=tips_df, x='tip', bins=20, kde = True)
```

## 2. `jointplot`:

```python
sns.jointplot(data = tips_df, x = 'total_bill', y = 'tip', kind = 'scatter', hue='day', palette='bright')
```

```python
sns.jointplot(data = tips_df, x = 'total_bill', y = 'tip', kind = 'reg')
```

## 3. `pairplot`:

```python
sns.pairplot(tips_df, hue='size', palette='rainbow')
```

## 4. `rugplot`:

```python
sns.rugplot(data=tips_df, x='tip', hue='sex')
```
