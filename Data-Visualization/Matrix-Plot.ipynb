# `Matrix plots`: 

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
flights_df=sns.load_dataset('flights')
flights_df
```

```python
tips_df=sns.load_dataset('tips')
tips_df
```

## ` Checking Data Co-relation`:

```python
bill_tip_size_df=tips_df[['total_bill', 'tip', 'size']]
```

```python
bill_tip_size_df
```

```python
bill_tip_size_df.corr()
```

## `Heat map`:

```python
sns.heatmap(bill_tip_size_df.corr(), annot= True)
```

## `Heatmap` by `Pivot Table`:

```python
flight_pvt_table = flights_df.pivot_table(index = 'month', columns = 'year', values = 'passengers', observed=True, aggfunc='mean')
```

```python
sns.heatmap(data=flight_pvt_table, annot=False)
```

## `Cluster map`: 

```python
sns.clustermap(bill_tip_size_df.corr())
```
