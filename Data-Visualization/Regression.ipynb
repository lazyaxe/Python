# `Regression Plots`:

```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
tips_df = sns.load_dataset("tips")
```

## There are mainly 2 two plots to see regression:
### 1. `jointplot`
### 2. `lmplot`

## `jointplot`

```python
sns.jointplot(data=tips_df, x='total_bill', y='tip', kind='reg')
```

## `lmplot()`

```python
sns.lmplot(data = tips_df, x = 'total_bill', y = 'tip')
```

```python
sns.lmplot(data = tips_df, x = 'total_bill', y = 'tip', hue = 'sex')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips_df, hue = 'sex', palette='mako')
```

> ### Changing markers and size of points/dots of lmplot():

```python
sns.lmplot(data = tips_df, x = 'total_bill', y = 'tip', hue = 'sex', palette='mako', markers=['^', '*'],
           scatter_kws={'s': 20}, legend=True)

#markers kwarg changes the shape/look of the points/dots on the plot, it can passed as a list for both values. Available shapes ['o', 'v', '^', '+', '*', '.', '<', '>']
#scatter_kws is the kwarg that changed the size of points/dots/markers on the plot
```

## Difference between `jointplot` and `lmplot`:
#### 1. `jointplot` is not just for regresion purposes, its versatile in many kinds of plots, while `lmplot` is speciallized for regression plots.
#### 2. jointplot doesn't provides as much customisation as lmplot for regeression, `lmplot` does.
#### 3. Using `hue=` or `palette=` in `kind='reg'` `jointplot` is not possible, while it is possible in `lmplot`.
#### 4. `jointplot` doesn't provide kwargs like `markers=` to change the shapes of points on plot or to change the size of plot by `scatter_kws=`.
### **Conclusion**: `lmplot` is better if you want to tweak the regression plots. `jointplot` is better for a quick view on how the regression looks.
