# `Regression Plots`:

```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
tips_df = sns.load_dataset("tips")
```

## There are mainly 2 two plots to see regression:
### 1. `jointplot`
### 2. `lmplot`

## 1. `jointplot`

```python
sns.jointplot(data=tips_df, x='total_bill', y='tip', kind='reg')
```

### 2. `lmplot()`

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips_df)
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips_df,hue = 'sex')
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips_df,hue = 'sex',palette='bright')
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips_df,hue = 'sex',palette='bright',markers=['o','v'],
           scatter_kws={'s':50})
```
