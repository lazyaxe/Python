# `Matplotlib` Basics:

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#Creating two sample vectors.
x = np.linspace(0,5,11)
x
```

```python
y = x **3 
y
```

## 1. `Basics` of `Plotting` 

```python
plt.plot(x,y)

plt.title("X vs Y")

plt.xlabel("X axis ———————>")

plt.ylabel("Y axis ———————>")
```

## 2. `Subplots` 🪟:
* ### Subplots are basically divisons in the canvas
* ### Subplot method has 3 parameters:
* ### First two parameters are the the size/shape of the canvas i.e. How many division you need(2x2=4 divisions, 2x3=6 divisions)?
* ### Last parameter is need to access the particular subplot. 

```python

plt.subplot(2, 2, 1)
plt.plot(x**3 ,y)

plt.subplot(2, 2, 2)
plt.plot(x*2, y)

plt.subplot(2, 2, 3)
plt.plot(x/2, y)

plt.subplot(2, 2, 4)
plt.plot(x, y/69420)
```

## Custom subplots:
* ### Matplotlib allows us to change the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.
* ### figsize is a tuple of the width and height of the figure in inches
* ### dpi is the dots-per-inch (pixel per inch).

```python
fig = plt.figure(figsize=(20, 8),dpi = 100)

"""
rect= [other axis width, other axis height, axis width, axis ]
"""
axis1 = fig.add_axes(rect=[0.1, -0.1, 0.2, 0.4])
axis1.plot(x, y)

axis2 = fig.add_axes(rect=[0.1, 0.4, 0.2, 0.4])
axis2.plot(y, x)
```

```python
from random import sample
data = sample(range(1,1000),100)
plt.hist(x=data, bins=10, color='orange')
plt.xlabel(xlabel="Numbers ———————>")
plt.ylabel(ylabel="Frequency ———————>")
```

## **Saving Plots** as **Image**

```python
fig = plt.figure()
ax = fig.add_axes(rect=[0.1, 0.1, 0.8, 0.8])
ax.plot(x,y)
plt.savefig('basicplot.png')
```
