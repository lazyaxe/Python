## Binomial Theorem/Bernoulli's Trials:
* Binomial theorem used a probability function, used to find probability of an event happening.
$$ P(A) = {{n}\choose{k}} \cdot p^k \cdot (1 - p)^{n-k}$$
where, 
* P(A) = The probablity of event A,
* n = no. of trials,
* p = probability of a success,
* k = no. of successfull trials,
${{n}\choose{k}}$ = combination of events

### Conditions:
1. Each event must be independent of previous event.
2. A event can only have 2 outcomes.
3. The number of trials must be finite.


```python
import math as m
import plotly.express as px
import pandas as pd
```

```python
def combination(n, r):
    return m.factorial(n) / ( m.factorial(n - r) * m.factorial(r))

def calculate_bt(n, p, k):
    return combination(n, k) * ((p ** k ) * ((1 - p) ** (n - k)))
```

## Binomial Distribution:
* #### Binomial distribution is a plot generated by changing a variable in binomial theorem.

There are two ways to plot it:
1. 'p' as variable
2. 'k' as variable

```python
#Draw a binomial distribution of flipping a coin 10 times
sample_space = ['heads', 'tails']
trials = 100
successfull_trials = 60 # 7 out of 10 are heads

def generate_list(number_of_trials):
    result = []
    success_probability = 0 # Assuming
    increment_constant = 1 / number_of_trials
    for i in range(trials):
        success_probability += increment_constant
        result.append(calculate_bt(n = trials, p = success_probability, k = successfull_trials))
    print(success_probability)
    return result

simulation_result = generate_list(trials)
simulation_result_df = pd.DataFrame(simulation_result)
```

```python
px.line(simulation_result_df)
```
