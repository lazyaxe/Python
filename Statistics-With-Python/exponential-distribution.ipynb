## Exponential Distribution:

### $ P(x) = \mu \cdot e^{- \mu t}$ , $\forall t \in [0, \infty)$ where, <br>
$\mu$ = expected time interval between events, $\mu$ = 1 / $\lambda$, <br>
t = time interval for an event or between events

### Conditions for exponential distribution:
1. Its's a PDF
2. Events must be independent.
3. Functionally, the exponential distribuion works in a opposite manner to Poisson's disribution.
4. Because it's a PDF, there's no *exact* probability i.e. $P(T = 15.\overline{0}s) = 0$

```python
import pandas as pd
import math as m
import plotly.express as px
import matplotlib.pyplot as plt
import sympy as sym
```

```python
def calculate_exponential(expected_mu, event):
    return expected_mu * m.e ** (- expected_mu * event)

def generate_list(expected_value):
    result = []
    i = 0
    counter = 0
    while True:
        result.append(calculate_exponential(expected_mu = expected_value, event = i))
        i += 0.01
        if counter == 6000:
            break
        counter += 1
    print(counter)
    return result

lambda_expected_value = 10
mu = 1 / lambda_expected_value

data = generate_list(expected_value = mu)
px.line(data)
```

```python
plt.plot(data)
```

## To find Probability i.e. area under the curve:
### $$ P( t2 \leqq T \leqq t1 ) = \mu \int_{\mathbf{t1}}^{\mathbf{t2}} e^{- \mu t} dt$$

```python
#calculating a probability of a region:
#example : find probablity of event occuring in t >= 65
t2 = 15
t1 = 0
t = sym.symbols('t')
sym.integrate(mu * m.e ** (- mu * t), (t, t1, t2))
```
