### Q1. A close match is defined as when either of the following is satisfied:
#### (i)   the match is "Tied" (captured in Column Q) 
#### (ii)  Balls remaining at the end of the match (Column V) < =  3
#### (iii) chasing team wins (Column Q) with  <= 1 Wickets remaining (Column R and S)  
#### (iv)  chasing team loses (Column Q) by <= 5 runs. (Column R and S)

```python
#(i) the match is "Tied" (captured in Column Q) 
cond_1 = (ipl_df['Winning_Team'] == 'Tied')

# (ii)  Balls remaining at the end of the match (Column V) < =  3
cond_2 = (ipl_df['Balls_Remaining'] <= 3)

# (iii) chasing team wins (Column Q) with  <= 1 Wickets remaining (Column R and S)
cond_3 = (ipl_df['Winning_Team'] == 'Chasing') & (ipl_df['Bat_Second_Wickets_Lost'] == 9)

# (iv)  chasing team loses (Column Q) by <= 5 runs. (Column R and S)
cond_4 = (ipl_df['Winning_Team'] == 'FirstBatting') & (ipl_df['Bat_First_Runs_Scored'] - ipl_df['Bat_Second_Runs_Scored'] <= 5)

#An either one or all can trigger a "close match" so it's a multi-or condition 
close_match_df = ipl_df[cond_1 | cond_2 | cond_3 | cond_4]['Year']
close_match_df.value_counts(sort = False)
```

## Q2. For the 4 teams listed: `Chennai`, `Hyderabad`, `Mumbai`, `Kolkata`, determine which team has won the most percentage of their matches batting first across all IPLs. To do this, compute for each team: 
### (i)   `A`=total number of matches won by the team 
### (ii)  `B`=how many matches they won batting first
### **Percent** won batting first = `(B / A) *100`

```python
teams = ['Chennai', 'Hyderabad', 'Mumbai', 'Kolkata']#list of team names

# (i) A = total matches won:
mask_A = ipl_df['Winner'].isin(values=teams)#equivalent to 'in' membership operator but for vectorized operations
A=ipl_df[mask_A]['Winner'].value_counts()

#(ii) B=how many matches they won batting first:
mask_B=(ipl_df['Winner']==ipl_df['Team_Batting_First']) & (ipl_df['Team_Batting_First'].isin(values=teams))
B=ipl_df[mask_B]['Winner'].value_counts()

percent = (B / A) * 100
percent.rename("Percent Batting First")
```

### b) Most consistent team in terms of lowest average wickets lost when batting first.

```python
#Why co-efficient of variation(cv)? : In stats cv is the standard measure of consistency and is calculated by
#c.v. = ((std deviation)/(mean))
#We multiply hundred to show it in percentage.

def co_effecient_of_variation(array_or_series):
    return np.std(array_or_series)/np.mean(array_or_series) * 100

wickets_lost = ipl_df[['Team_Batting_First', 'Bat_First_Wickets_Lost']]

wickets_lost['Team_Batting_First'].value_counts()#Finding names of all teams from 2008-2015, because some new teams could be introduced or some old teams can be discontinued...

team_name = ['Chennai', 'Hyderabad', 'Mumbai', 'Punjab', 'Kolkata', 'Rajasthan', 'Bangalore', 'Delhi', 'Pune', 'Kochi']

obj = wickets_lost.groupby(by='Team_Batting_First')

series = obj['Bat_First_Wickets_Lost'].apply(lambda x: x)
i = 0

consistency_of_teams = pd.DataFrame(data = team_name)
consistency_of_teams.rename(columns = {0: 'Team'}, inplace = True)

temp = []

for team in team_name:
    team_data=series[team_name[i]]
    temp.append(co_effecient_of_variation(team_data))
    i += 1

temp = pd.Series(data = temp)
consistency_of_teams = pd.concat([consistency_of_teams, temp], axis = 1)

consistency_of_teams.rename(columns = {0: 'Co-efficient of Variation of Wickets lost'}, inplace = True)

mask_of_consistency = consistency_of_teams['Co-efficient of Variation of Wickets lost'] == consistency_of_teams['Co-efficient of Variation of Wickets lost'].min()

consistency_of_teams[mask_of_consistency][['Team', 'Co-efficient of Variation of Wickets lost']]
```
